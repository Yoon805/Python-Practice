### Mission 
https://www.samsung.com/sec/ 사이트에 접속후 제품 내용 스크랩하기

```python
# 행과 열이 있는 데이터 처리 라이브러리
import pandas as pd
# 웹페이지 요청 라이브러리
import requests
# 페이지소스 이쁘게 가져오는 라이브러리
import bs4
```

```python
# 페이지에서 html 소스 가져오기
targetUrl = "https://www.samsung.com/sec/"
try:
    resp = requests.get(targetUrl)
except Exception as e:
    print(e)
    
html = resp.text
bs = bs4.BeautifulSoup(html, 'html.parser')
```

```python
# 목표 링크까지 Tag 줄여가기
divTag = bs.find(name = "div", attrs ={"class" :"footer-content" })

ulTag = divTag.find(name = "ul") 

liTag = ulTag.find(name="li")

ulTags2 = liTag.findAll(name = "ul")

targetATags1 = ulTags2[0].findAll(name ="a")

targetATags2 = ulTags2[1].findAll(name ="a")
```

```python
nameList = []
linkList = []

aTagsLen1 = len(targetATags1)

for i in range (0, aTagsLen1):
    #태그 텍스트 찾기
    tagTextName = targetATags1[i].text 
    #속성 찾기
    if targetATags1[i].attrs["href"][0] == "#":
        tagLinkInfo = targetUrl + targetATags1[i].attrs["href"]
    else:
        tagLinkInfo = targetATags1[i].attrs["href"]
    #태그명 찾기
    tagNameInfo = targetATags1[i].name
    nameList.append(tagTextName)
    linkList.append(tagLinkInfo)
```

```python
aTagsLen2 = len(targetATags1)

for i in range (0, aTagsLen2):
    #태그 텍스트 찾기
    tagTextName = targetATags2[i].text 
    #속성 찾기
    if targetATags2[i].attrs["href"][0] == "#":
        tagLinkInfo = targetUrl + targetATags2[i].attrs["href"]
    else:
        tagLinkInfo = targetATags2[i].attrs["href"]
    #태그명 찾기
    tagNameInfo = targetATags2[i].name
    nameList.append(tagTextName)
    linkList.append(tagLinkInfo)
```

```python
# 구한 값들을 zip 함수로 합치기
df = pd.DataFrame( zip( nameList, linkList) , columns = ["name","link"] )
```

```python
# csv 파일로 저장하기
df.to_csv("./practice.csv",index=False)
```

# 2. 링크정보만 수집
https://sparkkorea.com/%ed%80%b4%ec%a6%88/

```python
# 페이지에서 html 소스 가져오기
targetUrl = "https://sparkkorea.com/퀴즈/"
try:
    resp = requests.get(targetUrl)
except Exception as e:
    print(e)
    
html = resp.text
bs = bs4.BeautifulSoup(html, 'html.parser')

# 목표 링크까지 Tag 줄여가기
divTag = bs.find(name = "div", attrs ={"class" :"class_spark_quiz" })

aTags = divTag.findAll(name = "a")

nameList = []
linkList = []

aTagsLen = len(aTags)

for i in range (0, aTagsLen):
    #태그 텍스트 찾기
    tagTextName = aTags[i].text 
    #속성 찾기
    if aTags[i].attrs["href"][0] == "#":
        tagLinkInfo = targetUrl + aTags[i].attrs["href"]
    else:
        tagLinkInfo = aTags[i].attrs["href"]
    #태그명 찾기
    tagNameInfo = aTags[i].name
    nameList.append(tagTextName)
    linkList.append(tagLinkInfo)

# 구한 값들을 zip 함수로 합치기
df2 = pd.DataFrame( zip( nameList, linkList) , columns = ["name","link"] )
# csv 파일로 저장하기
df2.to_csv("./practice2.csv",index=False)
```
